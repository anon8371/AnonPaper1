```python
# correlation matrix between each of these? 
cosine_res = np.zeros((10,10))
shared_non_zeros = np.zeros((10,10))
eps = 1.0
for i in range(10):
    label_mask = train_l==i
    i_class_sum_vector = train_d[label_mask].sum(0) / label_mask.sum()
    
    for j in range(10):
        label_mask = train_l==j
        j_class_sum_vector = train_d[label_mask].sum(0) / label_mask.sum()
        
        cosine_res[i,j] = i_class_sum_vector/torch.norm(i_class_sum_vector)@j_class_sum_vector/torch.norm(j_class_sum_vector)
        #print(i, j, label_activations[i]/torch.norm(label_activations[i])@label_activations[j]/torch.norm(label_activations[j]))
        shared_non_zeros[i,j] = ((i_class_sum_vector>eps).type(torch.int) + (j_class_sum_vector>eps).type(torch.int)==2).sum()
cosine_res
```

```python
from types import SimpleNamespace
#from model_params import *
from dataset_params import *
from py_scripts.combine_params import *
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

load_path = "../experiments/AlexNetContLearningStarters/ReLU_ContLearning_AlexNetOnTop"
#"../experiments/AlexNetContLearningStarters/SDM_ContLearning_AlexNetOnTop_60PDead"
# 

checkpoint = torch.load(load_path, map_location=device)
params = checkpoint['hyper_parameters']
params = SimpleNamespace(**params)

extras = dict(
    epochs_to_train_for=10,
    num_workers=1, 
    adversarial_attacks=False, 
    device=device,
    dataset_path="../data/", 
    check_val_every_n_epoch=1,
    #k_approach = "GABA_SWITCH_ACT_BIN",
    log_metrics = False,
)

if load_path:
    print("LOADING IN A MODEL!!!")

params, model, data_module = get_params_net_dataloader(params.model_style, params.dataset, 
                                                       load_from_checkpoint=load_path, **extras)

model = model.to(device)
model.reload_neuron_activation_counters()
model.eval()
```

```python
data_module.setup(None)
val_loader = data_module.val_dataloader()
```

```python
label_activations = torch.zeros((10, params.nneurons[0]))
with torch.no_grad():
    for dind, d in enumerate(val_loader):
        x, y = d
        x, y = x.to(device), y.to(device)

        out = model.forward(x, output_model_data=params.validation_neuron_logger)
        if params.validation_neuron_logger:
            logits, model_data_dict = out
            
        acts = model_data_dict['post_acts']
            
        for ind in range(len(y)):
            label_activations[int(y[ind].cpu().numpy())] += torch.where(acts[ind]!=0, 1, 0).cpu()
```

```python
print("Number of neurons that are never activated for this input label")
for i in range(10):
    print(i, (label_activations[i]==0).sum(), "percentage", (label_activations[i]==0).sum()/len(label_activations[i]))
    print(i, (label_activations[i]<10).sum(), "percentage", (label_activations[i]<10).sum()/len(label_activations[i]))
    print("=====")
```

```python
# correlation matrix between each of these? 
res = np.zeros((10,10))
for i in range(10):
    for j in range(10):
        res[i,j] = label_activations[i]/torch.norm(label_activations[i])@label_activations[j]/torch.norm(label_activations[j])
        #print(i, j, label_activations[i]/torch.norm(label_activations[i])@label_activations[j]/torch.norm(label_activations[j]))
res
```
